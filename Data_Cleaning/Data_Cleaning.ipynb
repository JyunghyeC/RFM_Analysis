## 데이터 탐색

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rc
rc('font', family='AppleGothic')
```

```python
df = pd.read_csv('data 2.csv', encoding = 'unicode_escape')
df.head()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 결측치 삭제
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# InvoiceDate 타입 변환
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
# CustomerID 타입 변환
df['CustomerID'] = df['CustomerID'].astype('int64')
```

```python
df.info()
```

```python
df.shape
```

```python
df.head()
```

```python
# 중복 확인
df.duplicated().sum()
```

```python
# 중복 삭제
df.drop_duplicates(keep='first', inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.describe(include='all').T
```

```python
df.shape
```

```python
# UnitPrice가 0인 데이터 확인
df[df['UnitPrice'] == 0]
```

```python
# 모두 다른 상품과 다른 양, 날짜에 구매했는데도 불구하고 가격이 0으로 나옴
# Monetary 가 0으로 나올 수 있으니 삭제 결정
df = df[df['UnitPrice'] != 0]
```

```python
df[df['UnitPrice'] == 0]
```

```python
# InvoiceNo가 'C'로 시작하는 경우는 최소된 거래이기 때문에 삭제 결정
# 취소된 거래 확인
df[df['InvoiceNo'].str.contains('C')]
```

```python
df = df[~df['InvoiceNo'].str.startswith('C')]
```

```python
df.shape
```

```python
# 전처리된 데이터 저장
df.to_csv('rfm_data.csv', index=False)
```
